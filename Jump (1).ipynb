```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dadosjump.csv', sep=';')
df.head()
```

```python
#criaÃ§Ã£o colunas dependentes e idade
df['Idade'] = 2024 - df['Year_Birth']
df['Dependentes'] = df['Kidhome'] + df['Teenhome']
df.head()
```

##Compras por famÃ­lia

```python
df['Total_Compras'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

compras_por_filhos = df.groupby('Dependentes')['Total_Compras'].sum().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(df['Dependentes'], df['Total_Compras'], color='purple')
plt.title('RelaÃ§Ã£o entre NÃºmero de Filhos e Total de Compras')
plt.xlabel('NÃºmero de Filhos')
plt.ylabel('MÃ©dia do Total de Compras')
plt.grid(True)
plt.show()
```

```python
promo_data = df[['NumDealsPurchases', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']]

total_deals = promo_data.sum()

plt.figure(figsize=(10, 6))
plt.bar(total_deals.index, total_deals, color=['blue', 'green', 'orange', 'red', 'purple', 'brown'])
plt.title('Total de Compras com Desconto por Campanha de PromoÃ§Ã£o')
plt.xlabel('Campanha de PromoÃ§Ã£o')
plt.ylabel('Total de Compras com Desconto')
plt.xticks(rotation=45)
plt.legend(['Campanha 1', 'Campanha 2', 'Campanha 3', 'Campanha 4', 'Campanha 5', 'Ãšltima Campanha'])
plt.grid(True)
plt.show()
```

#Local de vendas x idade

```python
vendas_data = df[['Idade', 'NumWebPurchases', 'NumStorePurchases']]
soma_vendas_por_idade = vendas_data.groupby('Idade').sum().reset_index()


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(soma_vendas_por_idade['Idade'], soma_vendas_por_idade['NumWebPurchases'], color='blue')
plt.title('Vendas Online por Idade')
plt.xlabel('Idade')
plt.ylabel('NÃºmero de Compras Online')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.bar(soma_vendas_por_idade['Idade'], soma_vendas_por_idade['NumStorePurchases'], color='green')
plt.title('Vendas em Loja por Idade')
plt.xlabel('Idade')
plt.ylabel('NÃºmero de Compras em Loja')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
total_compras_online = df['NumWebPurchases'].sum()
total_compras_loja = df['NumStorePurchases'].sum()

if total_compras_online > total_compras_loja:
    print("HÃ¡ mais compras online do que em loja.")
elif total_compras_online < total_compras_loja:
    print("HÃ¡ mais compras em loja do que online.")
else:
    print("O nÃºmero de compras online e em loja Ã© o mesmo.")

labels = ['Compras Online', 'Compras em Loja']
sizes = [total_compras_online, total_compras_loja]
colors = ['blue', 'green']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('DistribuiÃ§Ã£o de Compras Online vs Compras em Loja')
plt.show()
```

#Dados pessoais x Vendas

```python
total_compras_clientes_que_reclamam = df[df['Complain'] == 1]['NumWebPurchases'].sum() + df[df['Complain'] == 1]['NumStorePurchases'].sum()
total_compras_clientes_que_nao_reclamam = df[df['Complain'] == 0]['NumWebPurchases'].sum() + df[df['Complain'] == 0]['NumStorePurchases'].sum()

if total_compras_clientes_que_reclamam < total_compras_clientes_que_nao_reclamam:
    print("Os clientes que reclamam compram menos.")
elif total_compras_clientes_que_reclamam > total_compras_clientes_que_nao_reclamam:
    print("Os clientes que reclamam nÃ£o necessariamente compram menos.")
else:
    print("NÃ£o hÃ¡ diferenÃ§a significativa no total de compras entre clientes que reclamam e clientes que nÃ£o reclamam.")

labels = ['Clientes que reclamam', 'Clientes que nÃ£o reclamam']
sizes = [total_compras_clientes_que_reclamam, total_compras_clientes_que_nao_reclamam]
colors = ['red', 'blue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('DistribuiÃ§Ã£o do Total de Compras entre Clientes que Reclamam e Clientes que nÃ£o Reclamam')
plt.show()
```
